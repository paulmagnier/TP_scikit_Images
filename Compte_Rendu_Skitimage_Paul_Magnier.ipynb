# Question 1

```python
fond_de_ciel = data.hubble_deep_field()
fond_de_ciel_crop = fond_de_ciel[0:512,0:512,:]
fond_de_ciel_crop_rouge = fond_de_ciel_crop[:,:,0]
plt.imshow(fond_de_ciel_crop_rouge, cmap='autumn')
```

```python
tf = numpy.fft.fft2(fond_de_ciel_crop_rouge)
re = tf.real
im = tf.imag
mod = abs(tf)
plt.imshow(mod)
```

```python
plt.subplot(131)
logre = numpy.log(re)
plt.imshow(logre)
plt.title("Log partie relle")
plt.subplot(132)
logim = numpy.log(im)
plt.imshow(logim)
plt.title("Log partie imaginaire")
plt.subplot(133)
logmod = numpy.log(mod)
plt.imshow(logmod)
plt.title("Log module")
```
